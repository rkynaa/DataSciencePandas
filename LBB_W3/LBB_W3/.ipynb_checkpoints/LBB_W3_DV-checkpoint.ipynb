## Task 1

Katakanlah kita akan melakukan analisis penjualan liquor pada 1 tahun terakhir(2018). Buatlah sebuah bar chart yang menampilkan lima kategori dengan jumlah penjualan tertinggi dalam satu tahun pada tahun tersebut. Kategori liquor apa yang **tidak termasuk** dalam kategori yang memiliki jumlah penjualan terbanyak, jika dilihat dari **kuantitas** penjualannya?

- [ ] Cocktails
- [ ] Liqueur
- [ ] Brandies
- [ ] Tequila
- [ ] Specialty Packages

*Tip:* `bottles_sold` memberikan informasi berapa botol liquor yang terjual dalam satu transaksi.

```python
# Answer
liquor_top5 = liquor.\
groupby('major_category_name').\
agg({'bottles_sold' : 'sum'}).\
sort_values('bottles_sold',ascending=True).\
tail(5)
liquor_top5.plot(kind="barh")
```

Lakukan eksplorasi lebih lanjut dari data `liquor.csv`. Akumulasikan dua pertanyaan bisnis dan cobalah untuk menemukan solusinya dengan metode EDA dari pandas!

```python
# Q1: Volume per bottle
liquor['yearly'] = liquor.date.dt.year
liquor_volume_bottle = liquor.\
groupby(['major_category_name','yearly']).\
agg({'sale_dollars': 'size'}).rename(
    columns= {'sale_dollars' : 'total_sale'}
)
liquor_volume_bottle.unstack().T.\
sort_values('yearly', ascending=False).\
plot(kind = "barh", stacked = True, figsize= (13,5)).\
legend(bbox_to_anchor=(1, 1))
```
