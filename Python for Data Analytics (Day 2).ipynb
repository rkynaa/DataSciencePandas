```python
employees = pd.DataFrame({
    'name': ['Anita', 'Brian'],
    'age': [34, 29],
    'joined': [pd.Timestamp('20190410'), pd.Timestamp('20171128')],
    'degree': [True, False],
    'hourlyrate': [35.5, 29],
    'division': ['HR', 'Product']
})
```

#### Types of data from `pandas`

1. Numeric:
    - `int64` = integer
    - `float64` = continuous (basically integer, but with decimal places)
2. Boolean `true / false`
3. datetime (Year-Month-Day / Year/Month/Day Hour:Minute:second)
4. Object
5. Category

> The difference between Object and Category is `object`-type data are more **"unique"** compared to `Cateogry`-type data

#### Change any data types into another data type

use `.astype()` and put any type of daya as its parameter

```python
employees['division'] = employees['division'].astype('category')
employees.dtypes
```

### `.describe()`

A function from `pandas` package to describe any data frame that are being targeted

```python
employees.describe()
```

#### `include=`

is an optional parameter that can be used for `describe()` function.

- `include='all'` is to include all parts of the data frame

```python
employees.describe(include='all')
```

- `include='(data type name)'` is to only use certain data type from the data frame

```python
employees.describe(include='datetime')
```
