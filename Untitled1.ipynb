```python
import pandas as pd
```

```python
household = pd.read_csv('dataanalysis/data_input/household.csv')
household.head()
```

```python
household.sub_category.value_counts()
```

```python
pd.crosstab(index=household['sub_category'],
           columns='count')
```

```python
pd.crosstab(index=household['sub_category'],
           columns=household['format'],
           normalize='index')
```

```python
pd.crosstab(index=household['sub_category'],
           columns=household['format'],
           margins=True,
           margins_name='Margin')
```

```python
pd.crosstab(index=household['sub_category'],
           columns=household['format'],
            values=household['unit_price'],
            aggfunc='mean',
           margins=True,
           margins_name='Margin')
```

```python
pd.crosstab(index=household['yearmonth'],
           columns=[household['format'],household['sub_category']],
           values=household['unit_price'],
           aggfunc='mean')
```

```python
pd.pivot_table(data=household,
              index='format',
              columns='format',
              aggfunc='size')
```

```python
pd.pivot_table(data=household,
              index='sub_category',
              columns='format',
              aggfunc='sum')
```

```python
pd.to_datetime('01-12-2019',format='%d-%m-%Y') == pd.to_datetime('01-12-2019',dayfirst=True)
```

```python
pd.pivot_table(data=household,
               index='sub_category',
               columns='format',
               values='unit_price',
               margins=True)
```
